```python
import matplotlib.pyplot as plt
from typing import Dict, List, Sequence, Optional, Literal
import numpy as np

def plot_user_metrics(
    data: Dict[int, Dict[str, Dict[str, float]]],
    recommender: str = "DemographicRecommender",
    metrics: Sequence[str] = ("precision", "recall", "f1"),
    users: Optional[Sequence[int]] = None,
    title: Optional[str] = None,
    top_n: Optional[int] = None,
    ylim: tuple = (0, 1),
    figsize: tuple = (8, 4)
) -> None:
    """
    Dibuja las mtricas por usuario para un recomendador concreto.

    Parameters
    ----------
    data : dict
        Diccionario con la estructura
        {user_id: {recommender_name: {metric: value, ...}, ...}, ...}
    recommender : str
        Nombre exacto del recomendador que se va a graficar.
    metrics : sequence of str
        Mtricas a representar (en el orden deseado).
    users : sequence of int, opcional
        Lista/tupla con los IDs de usuario en el orden que quieras mostrar.
        Si es None se usan todos los usuarios ordenados por ID.
    title : str, opcional
        Ttulo del grfico.  Por defecto: SR <nombre-sin-Recommender>.
    top_n : int, opcional
        Si lo proporcionas, se quedar solo con los *top n* usuarios
        segn el nmero de interacciones registradas en `data`.
        (Se toma el mismo orden que en *users* o por ID).
    ylim : tuple
        Lmites del eje Y.
    figsize : tuple
        Tamao de la figura en pulgadas.
    """
    # -------- 1. Seleccin y orden de usuarios -------------------------------
    if users is None:
        users = sorted(data.keys())
    if top_n is not None:
        users = users[:top_n]

    # Etiquetas en el eje X
    x_pos   = range(len(users))     
    x_labels = [f"User {u}" for u in users]

    # -------- 2. Recolectar las series de valores ---------------------------
    series = {m: [] for m in metrics}
    for u in users:
        rec_metrics = data.get(u, {}).get(recommender, {})
        for m in metrics:
            series[m].append(float(rec_metrics.get(m, 0)))  # 0 si falta dato

    # -------- 3. Dibujar -----------------------------------------------------
    plt.figure(figsize=figsize)
    
    # Colores fijos para recordar el ejemplo, pero puedes quitarlos
    default_colors = {
        "precision": "orange",
        "recall":    "firebrick",
        "f1":        "steelblue",
    }
    
    for m in metrics:
        plt.plot(
            x_pos,
            series[m],
            marker="o",
            linewidth=2,
            label=m.capitalize(),
            color=default_colors.get(m, None)  # deja que matplotlib elija si no
        )

    plt.xticks(x_pos, x_labels, rotation=45, ha="right")
    plt.ylim(*ylim)
    plt.ylabel("Valor")
    plt.title(title or f"SR {recommender.replace('Recommender', '').strip()}")
    plt.legend()
    plt.grid(axis="y", alpha=0.3)
    plt.tight_layout()
    plt.show()
```

```python
def plot_mae_per_user(
    data: Dict[int, Dict[str, Dict[str, float]]],
    recommenders: Optional[Sequence[str]] = None,
    users: Optional[Sequence[int]] = None,
    ylim: tuple = (0, None),          # None  ajuste automtico del mximo
    figsize: tuple = (8, 4),
    title: str | None = None,
    marker_size: int = 6
) -> None:
    """
    Dibuja el MAE por usuario para varios recomendadores.

    Parameters
    ----------
    data : dict
        Estructura {user_id: {recommender: {metric: value, ...}, ...}, ...}
    recommenders : sequence of str, opcional
        Lista de recomendadores a mostrar.  Si es None se usan todos los que
        aparezcan en `data`.
    users : sequence of int, opcional
        Orden de los IDs de usuario a mostrar.  Si None  ordenados por ID.
    ylim : tuple
        Lmites del eje Y.  (0, None) deja que matplotlib ajuste el mximo.
    figsize : tuple
        Tamao de la figura en pulgadas.
    title : str
        Ttulo del grfico.
    marker_size : int
        Tamao de los marcadores.
    """
    # ---------- 1. Conjunto de recomendadores y usuarios --------------------
    if recommenders is None:
        # Recolecta todos los nombres de recomendadores presentes
        rec_set = {r for u in data.values() for r in u.keys()}
        recommenders = sorted(rec_set)
    if users is None:
        users = sorted(data.keys())

    # Etiquetas del eje X
    x_pos    = range(len(users))
    x_labels = [str(u) for u in users]

    # ---------- 2. Construir las series de MAE ------------------------------
    mae_series = {rec: [] for rec in recommenders}

    for u in users:
        user_dict = data.get(u, {})
        for rec in recommenders:
            mae_val = user_dict.get(rec, {}).get("mae", float("nan"))
            mae_series[rec].append(float(mae_val))

    # ---------- 3. Dibujar ---------------------------------------------------
    plt.figure(figsize=figsize)

    for rec, values in mae_series.items():
        plt.plot(
            x_pos,
            values,
            marker="o",
            markersize=marker_size,
            linewidth=2,
            label=rec.replace("Recommender", "").strip()
        )

    plt.xticks(x_pos, x_labels, rotation=45, ha="right")
    plt.xlabel("ID de usuario")
    plt.ylabel("MAE")
    if ylim[1] is None:
        plt.ylim(ylim[0], max(max(v) for v in mae_series.values()) * 1.1)
    else:
        plt.ylim(*ylim)

    plt.title(title or "MAE por usuario y recomendador")
    plt.legend()
    plt.grid(axis="y", alpha=0.3)
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_metrics_per_recommender(
    agg_data,
    metrics=("precision", "recall", "f1", "mae"),
    figsize=(9, 5),
    title="Mtricas agregadas por recomendador",
    bar_width=0.18,
):
    """
    Dibuja un bar-chart agrupado donde cada barra representa
    el valor medio (o final) de una mtrica para cada recomendador.

    Parameters
    ----------
    agg_data : dict
        {'RecommenderName': {'precision': val, 'recall': val, ...}, ...}
    metrics : tuple[str]
        Mtricas a mostrar (deben existir como claves en cada sub-dict).
    figsize : tuple
        Tamao de la figura.
    title : str
        Ttulo del grfico.
    bar_width : float
        Anchura de cada barra.
    """
    recs = list(agg_data.keys())
    x = np.arange(len(recs))                # posiciones base

    plt.figure(figsize=figsize)

    for i, m in enumerate(metrics):
        vals = [float(agg_data[r][m]) for r in recs]
        plt.bar(
            x + i * bar_width,              # desplazamiento horizontal
            vals,
            width=bar_width,
            label=m.capitalize(),
        )

    plt.xticks(
        x + bar_width * (len(metrics) - 1) / 2,
        [r.replace("Recommender", "") for r in recs],
    )
    plt.ylabel("Valor")
    plt.title(title)
    plt.legend()
    plt.grid(axis="y", alpha=0.2)
    plt.tight_layout()
    plt.show()
```
