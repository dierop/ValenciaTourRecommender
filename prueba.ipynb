```python
import requests
import os
import urllib3
from src.data_loader import Data

api_key = open('api-key.txt').read()
lista_items = list(Data().items.name)

# Disable insecure request warning if you're aware of the risks
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)


url = "https://kgsearch.googleapis.com/v1/entities:search"
params = {
    'query': lista_items[6] + " España",
    'types': ['TouristAttraction', 'Thing', 'Place', 'EducationalOrganization', 'GovernmentOrganization', 'LocalBusiness',
              'Organization', 'Landmark'],
    'languages': "es",
    'limit': 1,
    'indent': True,
    'key': api_key,
}
print(params)

try:
    response = requests.get(url, params=params, verify=True)  # Enable verification for production
    response.raise_for_status()
    data = response.json()
    print(data)
except requests.exceptions.RequestException as e:
    print(f"Error making API request: {e}")

from IPython.display import display, Image
from urllib.request import urlopen
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from urllib.request import urlopen
from PIL import Image
import io

if len(data['itemListElement']) > 0:

    print(data['itemListElement'][0]['result']['name'])
    if 'image' in data['itemListElement'][0]['result']:
        response = urlopen(data['itemListElement'][0]['result']['image']['contentUrl'])
        image_data = response.read()
        image = Image.open(io.BytesIO(image_data))
        plt.imshow(image)
        plt.axis('off')  # Hide axes
        plt.show()
        print(data['itemListElement'][0]['result']['image']['contentUrl'])
    if 'url' in data['itemListElement'][0]['result']:
        print(data['itemListElement'][0]['result']['url'])
    if 'address' in data['itemListElement'][0]['result']:
        print(data['itemListElement'][0]['result']['address']['addressLocality'])
    if 'detailedDescription' in data['itemListElement'][0]['result']:
        print(data['itemListElement'][0]['result']['detailedDescription']['articleBody'])
    
    
```

```python
import requests

api_key = open('api-key-maps.txt').read()  # Your Google Maps API key

# Search for tourist attractions in Valencia
url = "https://maps.googleapis.com/maps/api/place/textsearch/json"
params = {
    'query': lista_items[0] + " in Valencia, Spain",
    'key': api_key,
    'language': 'es',  
}

try:
    response = requests.get(url, params=params)
    response.raise_for_status()
    data = response.json()
    
    for place in data.get('results', []):
        print(f"Name: {place.get('name')}")
        print(f"Direccion: {place.get('formatted_address')}")
        if 'opening_hours' in place:
            opening_status = place['opening_hours'].get('open_now')
            if opening_status is True:
                print("Abierto ahora: Sí")
            elif opening_status is False:
                print("Abierto ahora: No")
        else:
            print("Abierto ahora: Sin información")
        print(f"Valoracion en Google Maps: {place.get('rating', 'N/A')} sobre {place.get('user_ratings_total', 'N/A')} valoraciones")
        
except requests.exceptions.RequestException as e:
    print(f"Error: {e}")
```
