```python
import requests
api_key = open('api-key-maps.txt').read()  # Your Google Maps API key

def get_place_details(place_name, api_key):
    # Search for tourist attractions in Valencia
    url = "https://maps.googleapis.com/maps/api/place/textsearch/json"
    params = {
        'query': lista_items[0] + " in Valencia, Spain",
        'key': api_key,
        'language': 'es',  
    }

    try:
        response = requests.get(url, params=params)
        response.raise_for_status()
        data = response.json()
        
        for place in data.get('results', []):
            nombre = place.get('name')
            direccion = place.get('formatted_address')
            if 'opening_hours' in place:
                opening_status = place['opening_hours'].get('open_now')
                if opening_status is True:
                    #print("Abierto ahora: SÃ­")
                    abierto = 'SÃ­'
                elif opening_status is False:
                    #print("Abierto ahora: No")
                    abierto = 'No'
            else:
                #print("Abierto ahora: Sin informaciÃ³n")
                abierto = 'Sin informaciÃ³n'
            valoracion = place.get('rating', 'N/A')
            n_valoraciones = place.get('user_ratings_total', 'N/A')
            return nombre, direccion, abierto, valoracion, n_valoraciones
            
    except requests.exceptions.RequestException as e:
        return f"Error making API request: {e}"
```
